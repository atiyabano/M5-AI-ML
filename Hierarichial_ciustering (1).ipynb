```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering 
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
#Temprature data-- max & min
ax=np.array([[35,20],[40,23],[32,19],[50,30],[52,33],[24,12],[8,0],[7,-2],[3,0]])
```

```python
ax
```

```python
plt.scatter(ax[:,0],ax[:,1], color='blue')
plt.xlabel('Hiegest temp')
plt.ylabel('Lowest trmp')
plt.title('City temp')
plt.show()           
```

```python
a=AgglomerativeClustering(n_clusters=3)
```

```python
prediction=a.fit_predict(ax)
```

```python
prediction
```

```python
sns.scatterplot(x=ax[:,0],y=ax[:,1], hue=a, palette='viridis')
plt.xlabel('Hiegest temp')
plt.ylabel('Lowest temp')
plt.title('HC of temp')
plt.show()
```

```python
#dendogram 
li=linkage(ax,method='ward')
plt.figure(figsize=(5,4))
dendrogram(li,orientation='top',distance_sort='desecending',show_leaf_counts=True)
plt.title('Temprature')
plt.show()
```
