```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score # evaluation
```

```python
df = pd.read_csv('health_dataset.csv')
df
```

```python
df['PhysicalActivity']=df['PhysicalActivity'].map({'Low':0,'Medium':1,'High':2})
df['FamilyHistory']=df['FamilyHistory'].map({'No':0,'Yes':1})
```

```python
x=df.drop(columns='Outcome')
y=df['Outcome']
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
b=DecisionTreeClassifier(max_depth=4,random_state=4)
b
```

```python
b.fit(X_train, y_train)
```

```python
prediction=b.predict(X_test)
prediction
```

```python
accuracy=accuracy_score(y_test,prediction)
accuracy
```

```python
recall=recall_score(y_test,prediction)
recall
```

```python
precision=precision_score(y_test,prediction)
precision
```

```python
f1=f1_score(y_test,prediction)
f1
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8))
plot_tree(b,feature_names=x.columns,class_names=['No Diabetes','Diabetes'],filled=True)
plt.title("Decision Tree foe Diabeset")
plt.show()
```
